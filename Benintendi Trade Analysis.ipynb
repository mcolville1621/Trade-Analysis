```python
test = data1[['player_name','events', 'on_1b', 'on_2b', 'on_3b']]
HR_Judge = test.query('(events == "home_run") | (events == "single") | (events == "double")')
HR_Judge['no_runners'] = np.where((HR_Judge['on_1b'].isna()) & (HR_Judge['on_2b'].isna()) & (HR_Judge['on_3b'].isna()) , 1,0)
HR_Judge[['player_name', 'events', 'no_runners']].value_counts()
```
