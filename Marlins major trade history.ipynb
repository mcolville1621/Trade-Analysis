# Import packages

```python
from pybaseball import bwar_pitch
import pandas as pd
import numpy as np
from pybaseball import bwar_bat
```

# Start with pitching data

# Standardize Florida Teams and sort

```python
pitch = bwar_pitch()
pitch['team_ID'] = pitch['team_ID'].replace('FLA', 'MIA')
pitch = pitch.sort_values(['player_ID', 'year_ID', 'stint_ID'])
```

# Create fields to create if statments for Rookie and leaving Team identifiers

```python
pitch['prev_year'] = pitch['year_ID'].shift(1)
pitch['next_team'] = pitch['team_ID'].shift(-1)
pitch['prev_name'] = pitch['player_ID'].shift(1)
pitch['next_name'] = pitch['player_ID'].shift(-1)
pitch['next_year'] = pitch['year_ID'].shift(-1)
pitch['rookie'] = np.where(pitch['player_ID']==pitch['prev_name'], 'Not Rookie', 'Rookie')
pitch['left_team'] = np.where((pitch['player_ID']==pitch['next_name'])  
                              & (pitch['next_team']!=pitch['team_ID']), 'Left_team', 'NA')
pitch['mid_year'] = np.where((pitch['player_ID']==pitch['next_name'])  & (pitch['next_team']!='MIA') 
                             & (pitch['next_year']==pitch['year_ID']), 'traded', 'No')
pitch
```

# Filter Data for just MIA

```python
pitch1 = pitch.loc[(pitch['team_ID'] == 'MIA')]
pitch1
```

# Non-MIA Data

```python
No_Mia = pitch.loc[(pitch['team_ID'] != 'MIA')]
nomia_pivot = pd.pivot_table(No_Mia, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
nomia_pivot
```

# Create Pivot Tables for Career + MIA WAR and join together

```python
career_pitch = pd.pivot_table(pitch, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
nomia_pivot = pd.pivot_table(No_Mia, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
pitch_pivot = pd.pivot_table(pitch1, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
combined_1 = pitch_pivot.merge(career_pitch, on= 'player_ID', how = 'outer')
combined_1 = combined_1.rename(columns = {"WAR_x": "Marlins_WAR", "WAR_y": "Career_WAR"})
combined_1 = combined_1.loc[(combined_1['Marlins_WAR'] >= 2) | (combined_1['Career_WAR'] >= 2)]
```

# Join with original MIA data to get Rookie indicator and year and when the player left the team and the indicators for that

```python
rookie_pitch = combined_1.merge(pitch1, on= 'player_ID', how = 'left')
rookie_pitch1 = rookie_pitch.loc[(rookie_pitch['rookie'] == "Rookie")]
rookie_pitch1= rookie_pitch1.rename(columns = {"year_ID":"rookie_year"})
new_pitch= rookie_pitch1[['name_common', 'Career_WAR','Marlins_WAR', 'player_ID', 'rookie_year', 'team_ID', 'rookie']]
pitch2 = pitch1.loc[pitch1['left_team'] == 'Left_team']
pitch3 = pitch1.loc[pitch1['mid_year'] == 'traded']
tr_sign_pitch = new_pitch.merge(pitch2[['name_common', 'year_ID']], on= 'name_common', how = 'left')
tr_sign_pitch= tr_sign_pitch.rename(columns = {"year_ID":"year_left"})
tr_sign_pnew =tr_sign_pitch.merge(pitch3[['name_common', 'mid_year']], on= 'name_common', how = 'left') 
final_pitch = tr_sign_pnew.drop_duplicates().sort_values(['Marlins_WAR'], ascending = False)
final_pitch['Marlins_WAR1'] = final_pitch['Marlins_WAR'].astype('int64')
final_pitch
```

# compare 3 WAR

```python
final_pitch = final_pitch.merge(nomia_pivot, on= 'player_ID', how = 'left')
final_pitch = final_pitch.rename(columns = {"WAR": "NonMIA_WAR"})
final_pitch = final_pitch[['name_common', 'player_ID', 'team_ID', 'rookie_year', 'rookie', 'year_left', 'mid_year', 'Career_WAR', 'Marlins_WAR', 'NonMIA_WAR']]
final_pitch                              
```

# Indicates what kind of player they were

```python
final_pitch['all_star_type'] = np.where((final_pitch['Marlins_WAR'] < 3), "Good Player", 
    np.where((final_pitch['Marlins_WAR'] < 4) & (final_pitch['Marlins_WAR'] >= 3),"All Star", 
             np.where((final_pitch['Marlins_WAR'] >= 4), "Super Star", "NA")))
final_pitch
```

# Repeat for Bat Data - Format and create fields

```python
bat = bwar_bat()
bat['team_ID'] = bat['team_ID'].replace('FLA', 'MIA')
bat = bat.sort_values(['player_ID', 'year_ID', 'stint_ID'])
bat['prev_year'] = bat['year_ID'].shift(1)
bat['next_team'] = bat['team_ID'].shift(-1)
bat['prev_name'] = bat['player_ID'].shift(1)
bat['next_name'] = bat['player_ID'].shift(-1)
bat['next_year'] = bat['year_ID'].shift(-1)
bat['rookie'] = np.where(bat['player_ID']==bat['prev_name'], 'Not Rookie', 'Rookie')
bat['left_team'] = np.where((bat['player_ID']==bat['next_name'])  
                              & (bat['next_team']!=bat['team_ID']), 'Left_team', 'NA')
bat['mid_year'] = np.where((bat['player_ID']==bat['next_name'])  & (bat['next_team']!='MIA') 
                             & (bat['next_year']==bat['year_ID']), 'traded', 'No')
bat
```

# MIA Filter

```python
bat1 = bat.loc[(bat['team_ID'] == 'MIA')]
bat1
```

# Non-MIA Data

```python
No_Mia_b = bat.loc[(bat['team_ID'] != 'MIA')]
nomia_pivot_b = pd.pivot_table(No_Mia_b, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
nomia_pivot_b
```

# Pivot tables of WAR and joining to get Rookie and leaving MIA info

```python
career_pivot = pd.pivot_table(bat, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
bat_pivot = pd.pivot_table(bat1, values = 'WAR', index =['player_ID'], aggfunc = np.mean)
combined = bat_pivot.merge(career_pivot, on= 'player_ID', how = 'left')
combined = combined.rename(columns = {"WAR_x": "Marlins_WAR", "WAR_y": "Career_WAR"})
combined = combined.loc[(combined['Marlins_WAR'] >= 2) | (combined['Career_WAR'] >= 2)]
rookie_bat = combined.merge(bat1, on= 'player_ID', how = 'left')
rookie_bat1 = rookie_bat.loc[(rookie_bat['rookie'] == "Rookie")]
rookie_bat1= rookie_bat1.rename(columns = {"year_ID":"rookie_year"})
new_bat= rookie_bat1[['name_common', 'Career_WAR','Marlins_WAR', 'player_ID', 'rookie_year', 'team_ID', 'rookie']]
bat2 = bat1.loc[bat1['left_team'] == 'Left_team']
bat3 = bat1.loc[bat1['mid_year'] == 'traded']
tr_sign_bat = new_bat.merge(bat2[['name_common', 'year_ID']], on= 'name_common', how = 'left')
tr_sign_bat= tr_sign_bat.rename(columns = {"year_ID":"year_left"})
tr_sign_bnew =tr_sign_bat.merge(bat3[['name_common', 'mid_year']], on= 'name_common', how = 'left') 
final_bat = tr_sign_bnew.drop_duplicates().sort_values(['Marlins_WAR'], ascending = False)
final_bat['Marlins_WAR1'] = final_bat['Marlins_WAR'].astype('int64')
```

# compare 3 WAR

```python
final_bat = final_bat.merge(nomia_pivot_b, on= 'player_ID', how = 'left')
final_bat = final_bat.rename(columns = {"WAR": "NonMIA_WAR"})
final_bat = final_bat[['name_common', 'player_ID', 'team_ID', 'rookie_year', 'rookie', 'year_left', 'mid_year', 'Career_WAR', 'Marlins_WAR', 'NonMIA_WAR']]
final_bat                              
```

# Type of player indicators

```python
final_bat['all_star_type'] = np.where((final_bat['Marlins_WAR'] < 3), "Good Player", np.where((final_bat['Marlins_WAR'] < 4) & (final_bat['Marlins_WAR'] >= 3),
                                                    "All Star", np.where((final_bat['Marlins_WAR'] >= 4), "Super Star", "NA")))
batters = final_bat[['name_common', 'Career_WAR', 'Marlins_WAR', 'player_ID', 'rookie_year',
       'team_ID', 'rookie', 'year_left', 'mid_year', 'all_star_type']]
batters['player_type'] = 'batter'
final_bat
```

# Final Data

```python
pitchers = final_pitch[['name_common', 'Career_WAR', 'Marlins_WAR', 'player_ID', 'rookie_year',
       'team_ID', 'rookie', 'year_left', 'mid_year', 'all_star_type']]
pitchers['player_type'] = 'pitcher'
final = pd.concat([pitchers, batters])
final.sort_values(['Marlins_WAR'], ascending = False)
```
