```python
from pybaseball import team_batting, team_pitching, batting_stats, pitching_stats
import pandas as pd
```

```python
MLB = team_batting(2022)
Dv_lead = ('LAD', 'NYY', 'MIL', 'NYM', 'HOU', 'MIN')
WC_teams = ('TOR', 'SEA', 'TBR', 'ATL', 'SDP', 'STL')
Teams = ('BOS', 'BAL')
MLB = MLB.dropna(axis=1, how= 'all')
```

```python
two_teams = MLB.query('Team in @Teams')
two_teams_adv = two_teams[['Team',  'wOBA', 'wRC+', 'BABIP+', 'Barrel%', 'HardHit%', 'CSW%', 'O-Swing%', 'Z-Swing%', 'O-Contact%', 'Z-Contact%', 'Contact%', 'SwStr%', 'Pull%', 'Cent%', 'Oppo%', 'Hard%' ]]
two_teams_basic = two_teams[['Team', 'BsR', 'Def','OBP', 'OPS', 'ISO', 'HR', 'RBI', 'R', 'SO', 'SB', 'AVG']]
two_teams_adv
```

```python
Leaders_adv = MLB.query('Team in @Dv_lead')[['Team',  'wOBA', 'wRC+', 'BABIP+', 'Barrel%', 'HardHit%', 'CSW%', 'O-Swing%', 'Z-Swing%', 'O-Contact%', 'Z-Contact%', 'Contact%', 'SwStr%', 'Pull%', 'Cent%', 'Oppo%', 'Hard%' ]]
Leaders_adv1 = pd.DataFrame(Leaders_adv.mean()).T
Leaders_adv1['Team'] = 'Division Leaders'
WC_adv = MLB.query('Team in @WC_teams')[['Team',  'wOBA', 'wRC+', 'BABIP+', 'Barrel%', 'HardHit%', 'CSW%', 'O-Swing%', 'Z-Swing%', 'O-Contact%', 'Z-Contact%', 'Contact%', 'SwStr%', 'Pull%', 'Cent%', 'Oppo%', 'Hard%' ]]
WC_adv1 = pd.DataFrame(WC_adv.mean()).T
WC_adv1['Team'] = 'Wild Card Spots' 
```

```python
adv_merge = pd.concat([two_teams_adv,Leaders_adv1, WC_adv1])
adv_merge
```

```python
Leaders_basic = MLB.query('Team in @Dv_lead')[['Team', 'BsR', 'Def','OBP', 'OPS', 'ISO', 'HR', 'RBI', 'R', 'SO', 'SB', 'AVG']]
Leaders_basic1 = pd.DataFrame(Leaders_basic.mean()).T
Leaders_basic1['Team'] = 'Division Leaders'
WC_basic = MLB.query('Team in @WC_teams')[['Team', 'BsR', 'Def','OBP', 'OPS', 'ISO', 'HR', 'RBI', 'R', 'SO', 'SB', 'AVG']]
WC_basic1 = pd.DataFrame(WC_basic.mean()).T
WC_basic1['Team'] = 'Wild Card Spots'
basic_merge = pd.concat([two_teams_basic,Leaders_basic1, WC_basic1])
basic_merge
```

```python
Bat_comparison = basic_merge.merge(adv_merge, on= 'Team')
Bat_comparison. to_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\Deadline Analysis\Batting_AL_East.xlsx")
```
